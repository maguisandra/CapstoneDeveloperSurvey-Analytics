```python
import pandas as pd

# --- Load only the relevant columns for web frameworks ---
cols_webframe = ['ResponseId', 'WebframeHaveWorkedWith']
df_webframe = pd.read_csv(file_path, usecols=cols_webframe)

# --- Clean and normalize the separator ";" ---
# Remove extra spaces around semicolons and strip leading/trailing spaces
df_webframe['WebframeHaveWorkedWith'] = (
    df_webframe['WebframeHaveWorkedWith']
    .astype(str)
    .str.replace(r'\s*;\s*', ';', regex=True)  # ensure consistent semicolon separation
    .str.strip()  # remove leading/trailing spaces
)

# --- Split the column into lists of individual frameworks and explode ---
df_webframe = df_webframe.assign(
    WebframeHaveWorkedWith=df_webframe['WebframeHaveWorkedWith'].str.split(';')
).explode('WebframeHaveWorkedWith')

# --- Clean each value again (remove any remaining spaces) ---
df_webframe['WebframeHaveWorkedWith'] = df_webframe['WebframeHaveWorkedWith'].str.strip()

# --- Drop empty rows if any ---
df_webframe = df_webframe[df_webframe['WebframeHaveWorkedWith'] != '']

# --- Count frequency and get Top 10 frameworks ---
top10_webframes = df_webframe['WebframeHaveWorkedWith'].value_counts().head(10)

# --- Display the Top 10 results ---

# --- Optional: Save the cleaned dataframe ---
df_webframe.to_csv("df_webframe.csv", index=False)

# Display Top 10
print(top10_webframes)
```

###### The EdLevel column contains detailed degree names such as "Bachelor's degree (B.A., B.S., B.Eng., etc.)" or "Master's degree (M.A., M.S., M.Eng., MBA, etc.)".  To make analysis easier, we create a new column `dLevelSimplified by mapping each detailed degree to a simplified category: "Bachelor's", "Master's", "Professional", "Some college", or "Other".


```python
df["EdLevel"].value_counts()
```

```python
# Load only the relevant columns for Education Level
cols_ed = ['ResponseId', 'EdLevel']
df_EdLevel = pd.read_csv(file_path, usecols=cols_ed)

# Simplify education levels by keeping only the main degree name
# Simplify EdLevel column
def simplify_edlevel(x):
    if "Bachelor" in x:
        return "Bachelor's"
    elif "Master" in x:
        return "Master's"
    elif "Professional" in x:
        return "Professional"
    elif "Associate" in x:
        return "Associate"
    elif "Some college" in x:
        return "Some college"
    elif "Secondary school" in x or "High school" in x:
        return "Secondary"
    elif "Primary" in x or "elementary" in x:
        return "Primary"
    else:
        return "Other"

df_EdLevel['EdLevelSimplified'] = df_EdLevel['EdLevel'].apply(simplify_edlevel)

# Check the result
df_EdLevel['EdLevelSimplified'].value_counts()


# Apply the simplification
df_EdLevel['EdLevelSimplified'] = df_EdLevel['EdLevel'].apply(simplify_edlevel)

df_EdLevel['EdLevelSimplified'].value_counts()
```

###### We apply a complete mapping to standardize country names in our dataset.This allows us to group different variants of the same country under a single standard name (for example, "United States of America" becomes "USA").  



```python
# Load only ResponseId and Country
cols_country = ['ResponseId', 'Country']
df_country = pd.read_csv(file_path, usecols=cols_country)

# Strip spaces
df_country['Country'] = df_country['Country'].str.strip()

# Create a mapping dictionary for normalization
country_mapping = {
    "United States": "USA",
    "United States of America": "USA",
    "U.S.A.": "USA",
    "US": "USA",
    "United Kingdom of Great Britain and Northern Ireland": "UK",
    "United Kingdom": "UK",
    "Great Britain": "UK",
    "England": "UK",
    "Republic of India": "India",
    "India": "India",
    "Germany": "Germany",
    "Federal Republic of Germany": "Germany",
    "Canada": "Canada",
    "Australia": "Australia",
    "France": "France",
    "Brazil": "Brazil",
    "Russian Federation": "Russia",
    "Russia": "Russia",
    "China": "China",
    "People's Republic of China": "China",
    "Japan": "Japan",
    "South Korea": "South Korea",
    "Republic of Korea": "South Korea",
}
    
# Apply mapping
df_country['Country'] = df_country['Country'].replace(country_mapping)

# Check top countries
top_countries = df_country['Country'].value_counts().head(10)
print(top_countries)
```
